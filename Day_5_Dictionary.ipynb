```python
Keys=['shivalika','kiran','harsh']
values=['python', 'java','js']
```

here, we created 2 lists now in next part we'll see how do we convert them as dictionary.

```python
data=dict(zip(Keys,values))
data
```

this is how we created dictionary for lists or we can say for our easy access to any data.

```python
data['monica']='cs'
data
```

Like this we can alter dictionary.

```python
del data['harsh']
data
```
